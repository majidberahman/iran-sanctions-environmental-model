#Data Simulation

```python
# --- Step 1: Helper functions ---

import re
import numpy as np
import matplotlib.pyplot as plt

# Years of major sanction events
SANCTION_EVENTS = [
    (1979, "US asset freeze"),
    (1995, "US trade embargo"),
    (2006, "UN sanctions begin"),
    (2010, "UN sanctions expanded"),
    (2012, "EU oil embargo & SWIFT cutoff"),
    (2016, "JCPOA relief"),
    (2018, "US leaves JCPOA"),
    (2020, "US sanctions banks"),
    (2025, "UN snapback (E3)"),
]

def wb_to_long(df: pd.DataFrame) -> tuple[pd.DataFrame, str]:
    """Filter for Iran and reshape WB data to long (Year, Value)."""
    # detect year columns
    year_cols = [c for c in df.columns if re.fullmatch(r"\d{4}", str(c))]

    # filter Iran
    if "Country Code" in df.columns:
        iran_df = df[df["Country Code"] == "IRN"]
    else:
        iran_df = df[df["Country Name"].str.contains("Iran", case=False, na=False)]

    # melt to long
    long_df = iran_df.melt(
        id_vars=["Country Name","Country Code","Indicator Name","Indicator Code"],
        value_vars=year_cols,
        var_name="Year",
        value_name="Value"
    )
    long_df["Year"] = pd.to_numeric(long_df["Year"], errors="coerce")
    long_df["Value"] = pd.to_numeric(long_df["Value"], errors="coerce")
    long_df = long_df.dropna(subset=["Year","Value"]).sort_values("Year")

    indicator_name = iran_df["Indicator Name"].iloc[0]
    return long_df, indicator_name

def plot_with_sanctions(long_df, title):
    """Plot series with sanction vertical lines."""
    data = long_df.groupby("Year", as_index=False)["Value"].mean()
    plt.figure(figsize=(10,5))
    plt.plot(data["Year"], data["Value"], label="Iran")
    plt.plot(data["Year"], data["Value"].rolling(3, center=True, min_periods=1).mean(),
             linestyle="--", label="3-yr rolling mean")

    # add sanctions
    ymax = data["Value"].max()
    for year, label in SANCTION_EVENTS:
        if data["Year"].min() <= year <= data["Year"].max():
            plt.axvline(x=year, color="red", linestyle=":", linewidth=1)
            plt.text(year, ymax*1.02, str(year), rotation=90, ha="center", va="bottom", fontsize=8)

    plt.title(title)
    plt.xlabel("Year")
    plt.ylabel("Value")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# --- Step 2: Apply to each dataset ---

co2_long, co2_name = wb_to_long(co2_df)
plot_with_sanctions(co2_long, f"CO2 — {co2_name}")

water_long, water_name = wb_to_long(water_df)
plot_with_sanctions(water_long, f"Drinking Water — {water_name}")

forest_long, forest_name = wb_to_long(forest_df)
plot_with_sanctions(forest_long, f"Forest — {forest_name}")

electricity_long, electricity_name = wb_to_long(electricity_df)
plot_with_sanctions(electricity_long, f"Electricity — {electricity_name}")
```

# EDI Simulation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Assume the following DataFrames are already loaded in your environment:
# co2_df, water_df, forest_df, electricity_df

# --- 1. Preprocess and merge data to create the EDI index ---

def process_worldbank_data(df, indicator_name):
    """
    Cleans and reshapes a World Bank DataFrame for a specific country (Iran).
    """
    iran_data = df[df['Country Name'] == 'Iran, Islamic Rep.'].copy()
    if iran_data.empty:
        print(f"❌ Error: 'Iran, Islamic Rep.' data not found in the {indicator_name} file.")
        return pd.DataFrame()

    year_columns = [col for col in iran_data.columns if str(col).isdigit() and len(str(col)) == 4]

    processed_df = pd.melt(iran_data,
                           id_vars=['Country Name'],
                           value_vars=year_columns,
                           var_name='Year',
                           value_name='Value')

    processed_df['Year'] = pd.to_numeric(processed_df['Year'])
    processed_df['Value'] = pd.to_numeric(processed_df['Value'])
    processed_df['Indicator'] = indicator_name

    return processed_df.dropna()

print("✅ Starting data processing and EDI calculation...")

co2_clean = process_worldbank_data(co2_df, 'CO2 Emissions')
water_clean = process_worldbank_data(water_df, 'Drinking Water')
forest_clean = process_worldbank_data(forest_df, 'Forest Area')
electricity_clean = process_worldbank_data(electricity_df, 'Electricity Access')

all_data = pd.concat([co2_clean, water_clean, forest_clean, electricity_clean])

if all_data.empty:
    raise Exception("Combined data is empty. Please check your source files.")

edi_df = all_data.pivot_table(index='Year', columns='Indicator', values='Value').reset_index()

normalized_edi_df = edi_df.copy()

for ind in edi_df.columns[1:]:
    min_val = edi_df[ind].min()
    max_val = edi_df[ind].max()
    if max_val - min_val == 0:
        normalized_edi_df[ind] = 0.5
    else:
        normalized_edi_df[ind] = (edi_df[ind] - min_val) / (max_val - min_val)

normalized_edi_df['Drinking Water'] = 1 - normalized_edi_df['Drinking Water']
normalized_edi_df['Forest Area'] = 1 - normalized_edi_df['Forest Area']

weights = {
    'CO2 Emissions': 0.25,
    'Drinking Water': 0.25,
    'Forest Area': 0.25,
    'Electricity Access': 0.25
}

normalized_edi_df['EDI'] = 0
for ind, weight in weights.items():
    if ind in normalized_edi_df.columns:
        normalized_edi_df['EDI'] += weight * normalized_edi_df[ind]
    else:
        normalized_edi_df['EDI'] += weight * 0.5

print("\n📈 Calculated EDI Index (first few valid values):")
print(normalized_edi_df[['Year', 'EDI']].dropna().head())

# --- 2. Implement and simulate the multi-actor dynamic model ---
def edi_model(t, y, params):
    x, phi = y
    r0, r1, K, sanctions_impact = params['r0'], params['r1'], params['K'], params['sanctions_impact']

    dxdt = r0 + r1 * x - (x**2)/K + phi
    dphi_dt = sanctions_impact - (0.01 * phi)

    return [dxdt, dphi_dt]

# Find the first valid EDI value to start the simulation
first_valid_idx = normalized_edi_df['EDI'].first_valid_index()
if first_valid_idx is None:
    raise Exception("No valid EDI values found. Cannot run simulation.")

initial_edi = normalized_edi_df.loc[first_valid_idx, 'EDI']
initial_year = normalized_edi_df.loc[first_valid_idx, 'Year']
initial_friction = 0.1

t_span = (0, 30)
t_eval = np.linspace(t_span[0], t_span[1], 100)

# Run the simulations
sol_baseline = solve_ivp(
    edi_model, t_span, [initial_edi, initial_friction], args=(params_baseline,),
    dense_output=True, t_eval=t_eval
)

sol_sanctions = solve_ivp(
    edi_model, t_span, [initial_edi, initial_friction], args=(params_sanctions,),
    dense_output=True, t_eval=t_eval
)

# --- 3. Visualize the results ---
plt.figure(figsize=(12, 7))
plt.plot(sol_baseline.t + initial_year, sol_baseline.y[0], label='Baseline Scenario (No Sanctions)', color='blue', linewidth=2)
plt.plot(sol_sanctions.t + initial_year, sol_sanctions.y[0], label='Sanctions Scenario (Increased Institutional Friction)', color='red', linewidth=2)
plt.title('Simulation of Environmental Degradation Index (EDI) in Iran', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('EDI Value', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.ylim(0, 1.2)
plt.show()

print("\n✅ Simulation complete! The chart above shows the result.")
```

# Delay is Costly

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Assume the following DataFrames are already loaded in your environment:
# co2_df, water_df, forest_df, electricity_df

# --- Data Processing and EDI Calculation (same as before) ---
def process_worldbank_data(df, indicator_name):
    iran_data = df[df['Country Name'] == 'Iran, Islamic Rep.'].copy()
    if iran_data.empty:
        return pd.DataFrame()
    year_columns = [col for col in iran_data.columns if str(col).isdigit() and len(str(col)) == 4]
    processed_df = pd.melt(iran_data, id_vars=['Country Name'], value_vars=year_columns, var_name='Year', value_name='Value')
    processed_df['Year'] = pd.to_numeric(processed_df['Year'])
    processed_df['Value'] = pd.to_numeric(processed_df['Value'])
    processed_df['Indicator'] = indicator_name
    return processed_df.dropna()

co2_clean = process_worldbank_data(co2_df, 'CO2 Emissions')
water_clean = process_worldbank_data(water_df, 'Drinking Water')
forest_clean = process_worldbank_data(forest_df, 'Forest Area')
electricity_clean = process_worldbank_data(electricity_df, 'Electricity Access')
all_data = pd.concat([co2_clean, water_clean, forest_clean, electricity_clean])
if all_data.empty:
    raise Exception("Combined data is empty. Cannot perform simulation.")

edi_df = all_data.pivot_table(index='Year', columns='Indicator', values='Value').reset_index()
normalized_edi_df = edi_df.copy()
for ind in edi_df.columns[1:]:
    min_val = edi_df[ind].min()
    max_val = edi_df[ind].max()
    if max_val - min_val == 0:
        normalized_edi_df[ind] = 0.5
    else:
        normalized_edi_df[ind] = (edi_df[ind] - min_val) / (max_val - min_val)

normalized_edi_df['Drinking Water'] = 1 - normalized_edi_df['Drinking Water']
normalized_edi_df['Forest Area'] = 1 - normalized_edi_df['Forest Area']
weights = {'CO2 Emissions': 0.25, 'Drinking Water': 0.25, 'Forest Area': 0.25, 'Electricity Access': 0.25}
normalized_edi_df['EDI'] = sum(normalized_edi_df[ind] * weight for ind, weight in weights.items() if ind in normalized_edi_df.columns)

# Find first valid data point
first_valid_idx = normalized_edi_df['EDI'].first_valid_index()
if first_valid_idx is None:
    raise Exception("No valid EDI values found. Cannot run simulation.")
initial_edi = normalized_edi_df.loc[first_valid_idx, 'EDI']
initial_year = normalized_edi_df.loc[first_valid_idx, 'Year']
initial_friction = 0.1

# --- 2. Simulation for "Delay-is-Costly Law" ---
def edi_model_with_intervention(t, y, params):
    x, phi = y
    r0, r1, K, sanctions_impact, intervention_cost = params['r0'], params['r1'], params['K'], params['sanctions_impact'], params['intervention_cost']
    dxdt = r0 + r1 * x - (x**2)/K + phi - intervention_cost
    dphi_dt = sanctions_impact - (0.01 * phi)
    return [dxdt, dphi_dt]

# Define parameters and intervention costs
params_early = {'r0': 0.05, 'r1': 0.02, 'K': 1.0, 'sanctions_impact': 0.005, 'intervention_cost': 0.02}
params_late = {'r0': 0.05, 'r1': 0.02, 'K': 1.0, 'sanctions_impact': 0.005, 'intervention_cost': 0.02}

t_span = (0, 30)
t_eval = np.linspace(t_span[0], t_span[1], 100)

# Simulate early intervention (from the beginning)
sol_early = solve_ivp(
    edi_model_with_intervention, t_span, [initial_edi, initial_friction], args=(params_early,), dense_output=True, t_eval=t_eval
)

# Simulate late intervention (after 10 years)
t_early_phase = np.linspace(0, 10, 50)
sol_late_phase1 = solve_ivp(
    edi_model_with_intervention, (0, 10), [initial_edi, initial_friction], args=({'r0': 0.05, 'r1': 0.02, 'K': 1.0, 'sanctions_impact': 0.005, 'intervention_cost': 0},), dense_output=True, t_eval=t_early_phase
)
initial_late = sol_late_phase1.y[0][-1]
initial_phi_late = sol_late_phase1.y[1][-1]

t_late_phase = np.linspace(10, 30, 50)
sol_late_phase2 = solve_ivp(
    edi_model_with_intervention, (10, 30), [initial_late, initial_phi_late], args=(params_late,), dense_output=True, t_eval=t_late_phase
)

# Combine the two phases for late intervention
sol_late_t = np.concatenate((sol_late_phase1.t, sol_late_phase2.t))
sol_late_y = np.concatenate((sol_late_phase1.y[0], sol_late_phase2.y[0]))

# Plot the simulation
plt.figure(figsize=(12, 7))
plt.plot(sol_early.t + initial_year, sol_early.y[0], label='Early Intervention', color='green', linewidth=2)
plt.plot(sol_late_t + initial_year, sol_late_y, label='Delayed Intervention', color='orange', linewidth=2, linestyle='--')
plt.title('The "Delay-is-Costly" Law in Environmental Policy', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('EDI Value', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\n✅ 'Delay-is-Costly' simulation complete.")
print("The orange dashed line shows that delaying action leads to a higher EDI at all points,")
print("requiring more effort to reduce it in the future.")
```

# Sanctions as a Catalyst

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Data Processing and EDI Calculation (same as before)
def process_worldbank_data(df, indicator_name):
    iran_data = df[df['Country Name'] == 'Iran, Islamic Rep.'].copy()
    if iran_data.empty: return pd.DataFrame()
    year_columns = [col for col in iran_data.columns if str(col).isdigit() and len(str(col)) == 4]
    processed_df = pd.melt(iran_data, id_vars=['Country Name'], value_vars=year_columns, var_name='Year', value_name='Value')
    processed_df['Year'] = pd.to_numeric(processed_df['Year'])
    processed_df['Value'] = pd.to_numeric(processed_df['Value'])
    processed_df['Indicator'] = indicator_name
    return processed_df.dropna()

co2_clean = process_worldbank_data(co2_df, 'CO2 Emissions')
water_clean = process_worldbank_data(water_df, 'Drinking Water')
forest_clean = process_worldbank_data(forest_df, 'Forest Area')
electricity_clean = process_worldbank_data(electricity_df, 'Electricity Access')
all_data = pd.concat([co2_clean, water_clean, forest_clean, electricity_clean])
if all_data.empty: raise Exception("Combined data is empty. Cannot perform simulation.")

edi_df = all_data.pivot_table(index='Year', columns='Indicator', values='Value').reset_index()
normalized_edi_df = edi_df.copy()
for ind in edi_df.columns[1:]:
    min_val = edi_df[ind].min()
    max_val = edi_df[ind].max()
    if max_val - min_val == 0: normalized_edi_df[ind] = 0.5
    else: normalized_edi_df[ind] = (edi_df[ind] - min_val) / (max_val - min_val)
normalized_edi_df['Drinking Water'] = 1 - normalized_edi_df['Drinking Water']
normalized_edi_df['Forest Area'] = 1 - normalized_edi_df['Forest Area']
weights = {'CO2 Emissions': 0.25, 'Drinking Water': 0.25, 'Forest Area': 0.25, 'Electricity Access': 0.25}
normalized_edi_df['EDI'] = sum(normalized_edi_df[ind] * weight for ind, weight in weights.items() if ind in normalized_edi_df.columns)

first_valid_idx = normalized_edi_df['EDI'].first_valid_index()
if first_valid_idx is None: raise Exception("No valid EDI values found. Cannot run simulation.")
initial_edi = normalized_edi_df.loc[first_valid_idx, 'EDI']
initial_year = normalized_edi_df.loc[first_valid_idx, 'Year']
initial_friction = 0.1

# --- 2. Simulation for Sanctions as a Catalyst ---
def edi_model_with_shock(t, y, params):
    x, phi = y
    r0, r1, K, sanctions_impact = params['r0'], params['r1'], params['K'], params['sanctions_impact']

    # Apply a shock after 10 years to simulate a new round of sanctions
    shock_impact = 0.02 if t >= 10 else 0

    dxdt = r0 + r1 * x - (x**2)/K + phi
    dphi_dt = sanctions_impact + shock_impact - (0.01 * phi)

    return [dxdt, dphi_dt]

# Parameters for baseline and shock scenarios
params_baseline = {'r0': 0.05, 'r1': 0.02, 'K': 1.0, 'sanctions_impact': 0.005}
params_shock = {'r0': 0.05, 'r1': 0.02, 'K': 1.0, 'sanctions_impact': 0.005}

t_span = (0, 30)
t_eval = np.linspace(t_span[0], t_span[1], 100)

# Run simulations
sol_baseline = solve_ivp(
    edi_model, t_span, [initial_edi, initial_friction], args=(params_baseline,), dense_output=True, t_eval=t_eval
)

sol_shock = solve_ivp(
    edi_model_with_shock, t_span, [initial_edi, initial_friction], args=(params_shock,), dense_output=True, t_eval=t_eval
)

# Plot the simulation
plt.figure(figsize=(12, 7))
plt.plot(sol_baseline.t + initial_year, sol_baseline.y[0], label='Baseline Scenario (Constant Sanctions)', color='blue', linewidth=2)
plt.plot(sol_shock.t + initial_year, sol_shock.y[0], label='Sanctions Shock Scenario', color='red', linewidth=2, linestyle='--')
plt.title('Sanctions as a Catalyst for Environmental Degradation', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('EDI Value', fontsize=12)
plt.axvline(x=initial_year + 10, color='gray', linestyle=':', linewidth=1.5, label='Sanctions Shock Begins (Year 10)')
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\n✅ 'Sanctions as a Catalyst' simulation complete.")
print("The red dashed line shows how an additional shock (representing a new round of sanctions)")
print("accelerates the rate of environmental degradation relative to the baseline.")
```

# Impact of Gradual Sanctions Removal

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Assume the DataFrames are already loaded
# co2_df, water_df, forest_df, electricity_df

# --- Data Processing and EDI Calculation (same as before) ---
def process_worldbank_data(df, indicator_name):
    iran_data = df[df['Country Name'] == 'Iran, Islamic Rep.'].copy()
    if iran_data.empty: return pd.DataFrame()
    year_columns = [col for col in iran_data.columns if str(col).isdigit() and len(str(col)) == 4]
    processed_df = pd.melt(iran_data, id_vars=['Country Name'], value_vars=year_columns, var_name='Year', value_name='Value')
    processed_df['Year'] = pd.to_numeric(processed_df['Year'])
    processed_df['Value'] = pd.to_numeric(processed_df['Value'])
    processed_df['Indicator'] = indicator_name
    return processed_df.dropna()

co2_clean = process_worldbank_data(co2_df, 'CO2 Emissions')
water_clean = process_worldbank_data(water_df, 'Drinking Water')
forest_clean = process_worldbank_data(forest_df, 'Forest Area')
electricity_clean = process_worldbank_data(electricity_df, 'Electricity Access')
all_data = pd.concat([co2_clean, water_clean, forest_clean, electricity_clean])
if all_data.empty: raise Exception("Combined data is empty. Cannot perform simulation.")

edi_df = all_data.pivot_table(index='Year', columns='Indicator', values='Value').reset_index()
normalized_edi_df = edi_df.copy()
for ind in edi_df.columns[1:]:
    min_val = edi_df[ind].min()
    max_val = edi_df[ind].max()
    if max_val - min_val == 0: normalized_edi_df[ind] = 0.5
    else: normalized_edi_df[ind] = (edi_df[ind] - min_val) / (max_val - min_val)
normalized_edi_df['Drinking Water'] = 1 - normalized_edi_df['Drinking Water']
normalized_edi_df['Forest Area'] = 1 - normalized_edi_df['Forest Area']
weights = {'CO2 Emissions': 0.25, 'Drinking Water': 0.25, 'Forest Area': 0.25, 'Electricity Access': 0.25}
normalized_edi_df['EDI'] = sum(normalized_edi_df[ind] * weight for ind, weight in weights.items() if ind in normalized_edi_df.columns)

first_valid_idx = normalized_edi_df['EDI'].first_valid_index()
if first_valid_idx is None: raise Exception("No valid EDI values found. Cannot run simulation.")
initial_edi = normalized_edi_df.loc[first_valid_idx, 'EDI']
initial_year = normalized_edi_df.loc[first_valid_idx, 'Year']
initial_friction = 0.1

# --- 2. Simulation for Sanctions Transition Effect ---
def edi_model_with_transition(t, y, params):
    x, phi = y
    r0, r1, K = params['r0'], params['r1'], params['K']

    # Sanctions impact function (decreases over time after 10 years)
    sanctions_impact = 0.005
    if t > 10:
        sanctions_impact = 0.005 * np.exp(-0.1 * (t - 10)) # Exponential decay

    dxdt = r0 + r1 * x - (x**2)/K + phi
    dphi_dt = sanctions_impact - (0.01 * phi)

    return [dxdt, dphi_dt]

params_baseline = {'r0': 0.05, 'r1': 0.02, 'K': 1.0, 'sanctions_impact': 0.005}
params_transition = {'r0': 0.05, 'r1': 0.02, 'K': 1.0}

t_span = (0, 30)
t_eval = np.linspace(t_span[0], t_span[1], 100)

sol_baseline = solve_ivp(
    edi_model, t_span, [initial_edi, initial_friction], args=(params_baseline,), dense_output=True, t_eval=t_eval
)

sol_transition = solve_ivp(
    edi_model_with_transition, t_span, [initial_edi, initial_friction], args=(params_transition,), dense_output=True, t_eval=t_eval
)

# Plot the simulation with improved visualization
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(14, 8))

# Plot the simulation results
plt.plot(sol_baseline.t + initial_year, sol_baseline.y[0], label='Baseline (Constant Sanctions)', color='salmon', linewidth=3)
plt.plot(sol_transition.t + initial_year, sol_transition.y[0], label='Transition from Sanctions', color='teal', linewidth=3, linestyle='--')

# Plot the real data points
real_data = normalized_edi_df.dropna(subset=['EDI'])
plt.scatter(real_data['Year'], real_data['EDI'], color='k', zorder=10, label='Real Data Points', s=50)

# Add annotations
plt.axvline(x=initial_year + 10, color='gray', linestyle=':', linewidth=2, label='Sanctions Begin to Decrease (Year 10)')
plt.title('Impact of Gradual Sanctions Removal on Environmental Degradation', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=14, labelpad=15)
plt.ylabel('Environmental Degradation Index (EDI)', fontsize=14, labelpad=15)
plt.legend(fontsize=12, loc='upper left')
plt.ylim(0, 1.2)
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()

print("\n✅ 'Sanctions Transition Effect' simulation complete.")
print("The dotted line at year 10 marks the beginning of sanctions removal.")
print("The plot shows how a gradual lifting of sanctions can lead to a long-term improvement in the EDI.")
```

# System Stability and Resilience

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Assume the following DataFrames are already loaded in your environment:
# co2_df, water_df, forest_df, electricity_df

# --- Data Processing and EDI Calculation (same as before) ---
def process_worldbank_data(df, indicator_name):
    iran_data = df[df['Country Name'] == 'Iran, Islamic Rep.'].copy()
    if iran_data.empty: return pd.DataFrame()
    year_columns = [col for col in iran_data.columns if str(col).isdigit() and len(str(col)) == 4]
    processed_df = pd.melt(iran_data, id_vars=['Country Name'], value_vars=year_columns, var_name='Year', value_name='Value')
    processed_df['Year'] = pd.to_numeric(processed_df['Year'])
    processed_df['Value'] = pd.to_numeric(processed_df['Value'])
    processed_df['Indicator'] = indicator_name
    return processed_df.dropna()

co2_clean = process_worldbank_data(co2_df, 'CO2 Emissions')
water_clean = process_worldbank_data(water_df, 'Drinking Water')
forest_clean = process_worldbank_data(forest_df, 'Forest Area')
electricity_clean = process_worldbank_data(electricity_df, 'Electricity Access')
all_data = pd.concat([co2_clean, water_clean, forest_clean, electricity_clean])
if all_data.empty: raise Exception("Combined data is empty. Cannot perform simulation.")

edi_df = all_data.pivot_table(index='Year', columns='Indicator', values='Value').reset_index()
normalized_edi_df = edi_df.copy()
for ind in edi_df.columns[1:]:
    min_val = edi_df[ind].min()
    max_val = edi_df[ind].max()
    if max_val - min_val == 0: normalized_edi_df[ind] = 0.5
    else: normalized_edi_df[ind] = (edi_df[ind] - min_val) / (max_val - min_val)
normalized_edi_df['Drinking Water'] = 1 - normalized_edi_df['Drinking Water']
normalized_edi_df['Forest Area'] = 1 - normalized_edi_df['Forest Area']
weights = {'CO2 Emissions': 0.25, 'Drinking Water': 0.25, 'Forest Area': 0.25, 'Electricity Access': 0.25}
normalized_edi_df['EDI'] = sum(normalized_edi_df[ind] * weight for ind, weight in weights.items() if ind in normalized_edi_df.columns)

first_valid_idx = normalized_edi_df['EDI'].first_valid_index()
if first_valid_idx is None: raise Exception("No valid EDI values found. Cannot run simulation.")
initial_edi = normalized_edi_df.loc[first_valid_idx, 'EDI']
initial_year = normalized_edi_df.loc[first_valid_idx, 'Year']
initial_friction = 0.1

# --- 2. Simulation for System Stability and Resilience ---
def edi_model_resilience(t, y, params):
    x, phi = y
    r0, r1, K, sanctions_impact = params['r0'], params['r1'], params['K'], params['sanctions_impact']

    # Intervention force to push EDI down towards a stable, non-zero level
    intervention = -0.015 * x

    dxdt = r0 + r1 * x - (x**2)/K + phi + intervention
    dphi_dt = sanctions_impact - (0.01 * phi)

    return [dxdt, dphi_dt]

# Set model parameters
params_stable = {'r0': 0.05, 'r1': 0.02, 'K': 1.0, 'sanctions_impact': 0.005}
params_shocked = {'r0': 0.05, 'r1': 0.02, 'K': 1.0, 'sanctions_impact': 0.01}

t_span = (0, 30)
t_eval = np.linspace(t_span[0], t_span[1], 100)

# Run simulations for two scenarios
sol_stable = solve_ivp(
    edi_model_resilience, t_span, [initial_edi, initial_friction], args=(params_stable,), dense_output=True, t_eval=t_eval
)

sol_shocked = solve_ivp(
    edi_model_resilience, t_span, [initial_edi, initial_friction], args=(params_shocked,), dense_output=True, t_eval=t_eval
)

# Plot the simulation with improved visualization
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(14, 8))

# Plot the simulation results
plt.plot(sol_stable.t + initial_year, sol_stable.y[0], label='Stable Scenario (Lower Friction)', color='darkblue', linewidth=3)
plt.plot(sol_shocked.t + initial_year, sol_shocked.y[0], label='Shocked Scenario (Higher Friction)', color='red', linewidth=3, linestyle='--')

# Plot the real data points
real_data = normalized_edi_df.dropna(subset=['EDI'])
plt.scatter(real_data['Year'], real_data['EDI'], color='k', zorder=10, label='Real Data Points', s=50)

plt.title('System Stability and Resilience of Environmental Degradation Index', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Year', fontsize=14, labelpad=15)
plt.ylabel('Environmental Degradation Index (EDI)', fontsize=14, labelpad=15)
plt.legend(fontsize=12, loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
